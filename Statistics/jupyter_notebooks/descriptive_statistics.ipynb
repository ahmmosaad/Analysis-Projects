### **Center Tendancy**

Mean, Median, Mode

```python
aqi = df['aqi']

aqi.mean()
```

```python
median = aqi.median()
median
```

mode = df['state_name'].mode()

### **Dispersion**

Range - Standard Division - Variance

```python
range = aqi.max() - aqi.min()
range
```

```python
standard_division = aqi.std()
standard_division
```

```python
variance = standard_division**2
variance
```

### **Position**

Percentile - Quantile - IQR - 5 number summary

```python
percntile_10 = np.percentile(aqi, 10)
percntile_10
```

```python
Q1 = np.percentile(aqi, 25)
Q2 = np.percentile(aqi, 50)
Q3 = np.percentile(aqi, 75)
Q1 , Q2 , Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
min = aqi.min()
max = aqi.max()

min, max, Q1, Q2 , Q3 
```
