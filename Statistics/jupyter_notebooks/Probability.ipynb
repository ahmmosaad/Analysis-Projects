### Emprical Rule of Normal Distribution

stating that approximately 68% of data falls within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations

```python
mean = df['OVERALL_LI'].mean()
std = df['OVERALL_LI'].std()
mean , std
```

```python
the_68_percentage_left = mean - (1*std)  
the_68_percentage_right = mean + (1*std)  
the_68_percentage_left , the_68_percentage_right
```

```python
the_95_percentage_left = mean - (2*std)  
the_95_percentage_right = mean + (2*std)  
the_95_percentage_left , the_95_percentage_right
```

```python
the_99_percentage_left = mean - (3*std)  
the_99_percentage_right = mean + (3*std)  
the_99_percentage_left , the_99_percentage_right
```

```python
 plt.hist(df['OVERALL_LI'], bins = 20)

plt.axvline(mean, color="white")
plt.axvline(the_68_percentage_left, color="green",ymax=0.75, label="68% lies here")
plt.axvline(the_68_percentage_right, color="green", ymax=0.75,)

 
plt.axvline(the_95_percentage_left, color="red",ymax=0.60, label="95% lies here")
plt.axvline(the_95_percentage_right, color="red", ymax=0.60,)


plt.axvline(the_99_percentage_left, color="yellow", ymax=0.30,  label="99.7% lies here")
plt.axvline(the_99_percentage_right, color="yellow" ,ymax=0.30,)

plt.legend(loc="upper right")
plt.show()
```

**Z Score**

Indicates how many standard deviations a data point is above or below the mean of a dataset

```python
dd = df["OVERALL_LI"]
stats.zscore(dd)
```
